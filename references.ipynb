```python
import gzip
import pandas as pd

# Define the path to the .tsv.gz file
tsv_gz_file_path = 'data/hBreast/cell_feature_matrix/features.tsv.gz'  # Replace with the path to your .tsv.gz file

# Open and read the .tsv.gz file
with gzip.open(tsv_gz_file_path, 'rt') as file:
    # Read the file into a pandas DataFrame
    df = pd.read_csv(file, delimiter='\t')
    
    # Display the first few rows of the DataFrame
    print(df.head(25))
```

```python
import gzip
import pandas as pd
from scipy.io import mmread

# Define the path to the .mtx.gz file
mtx_gz_file_path = 'data/hBreast/cell_feature_matrix/matrix.mtx.gz'  # Replace with the path to your .mtx.gz file

# Open and read the .mtx.gz file
with gzip.open(mtx_gz_file_path, 'rt') as file:
    # Read the file into a sparse matrix
    sparse_matrix = mmread(file)
    
    # Convert the sparse matrix to a dense format (if needed)
    dense_matrix = sparse_matrix.toarray()
    
    # Display the shape of the matrix
    print("Matrix shape:", dense_matrix.shape)
    
    # Display the first few rows of the matrix
    print(dense_matrix[:5, :5])
```
