```python
import tifffile as tf
import matplotlib.pyplot as plt

# Replace 'path_to_image.ome.tif' with the path to your OME-TIFF file
image_path = 'data/hBreast/morphology.ome.tif'

# Use tifffile to read the OME-TIFF file
with tf.TiffFile(image_path) as tif:
    # Assuming the first dimension is the channel or z-stack,
    # we select the first one for visualization
    image = tif.asarray()[0]  # This selects the first channel or z-stack

# Display the image using matplotlib
plt.imshow(image, cmap='Blues') # You can change the color map if needed
plt.axis('off') # Hide the axis
plt.show()
```
