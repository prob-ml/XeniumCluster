```python
spot_sizes = [50, 75, 100]
in_billions = 1_000_000_000
method="BayXenSmooth"
init_methods = ["mclust", "K-Means", "Leiden"]
PC_list = [3, 5, 10, 15, 25]
```

```python
import os
import json
import numpy as np

# Define the directory where the results are stored
results_dir = "results/hBreast/BayXenSmooth/mpd"

# Loop through all subdirectories in the results directory
min_mpd = float('inf')
min_filename = ""
for root, dirs, files in os.walk(results_dir):
    for file in files:
        # Check if the file is named morans_i_by_gene.json
        if file.endswith(".json"):
            filename = os.path.join(root, file)
            if os.path.exists(filename) and "sum" not in filename:
                with open(filename, "r") as mpd_dict:
                    current_mpd = json.load(mpd_dict)
                    # Check if the sum of current_mpd.values() is not NaN
                    if not any(np.isnan(value) for value in current_mpd.values()):
                        print(filename, sum(current_mpd.values()) / in_billions )
                        if sum(current_mpd.values()) / in_billions < min_mpd:
                            min_mpd = sum(current_mpd.values()) / in_billions 
                            min_filename = filename      
```
